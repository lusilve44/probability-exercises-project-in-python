### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import numpy as np
import pandas as pd

#Funtion to roll the dice.

np.random.seed(42)

def roll_2_dices(times): #roll the dices a certain amount of times, evaluates the sum and determines if the conditions are true
    conditions_were_true = 0
    for _ in range(times):
        result_of_dices = np.random.randint(1,7,2)
        result = sum(result_of_dices)
        if (result%2 == 0) or (result > 7):
            conditions_were_true += 1
    return (conditions_were_true / times)

print(f'The probability of rolling two dices and the sum is even or greater than seven is: {roll_2_dices(1000)*100}%')
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#import libraries
import numpy as np
import pandas as pd

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i > 9 and i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'  

#Initialize important variables

np.random.seed(42)

n_simulations = 1000
white3_or_red2 = 0
same_colour = 0
white3_and_red2 = 0

for i in range(n_simulations):
    
    #make a list of the colors that we choose
    colours = np.random.randint(0,60,5)

    #Convert it to a numpy array
    colours = np.array(colours)

    #find the number of each that we picked
    white = 0
    red = 0
    green = 0
    for i in colours:
        if i < 10:
            white += 1
        elif i > 9 and i < 30:
            red += 1
        else:
            green += 1
    
    #Keep track if the combination met the criteria
    if (white == 3) or (red == 2):
        white3_or_red2 += 1
    elif (white == 5) or (red == 5) or (green == 5):
        same_colour += 1
    
    if (white == 3) and (red == 2):
        white3_and_red2 += 1

probability_1 = white3_or_red2 / n_simulations
probability_2 = same_colour / n_simulations
probability_3 = white3_and_red2 / n_simulations

print(f'the probability of 3 white or 2 red is {probability_1*100}%')
print(f'the probability of all the same color is {probability_2*100}%')
print(f'the probability of 3 white and 2 red is {probability_3*100}%')
```
